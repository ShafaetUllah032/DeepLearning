# Handling Date and Time Variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Date Handling

```python
data=pd.read_csv('../data/order.csv')
```

```python
data.head()
```

```python
data['date'].info()
```

```python
data['date'].describe()
```

```python
data=data[['date']]
```

```python
data['date']=pd.to_datetime(data['date'])

data['date'].sample(5)
```

### 1. Extract year

```python
data['year']=data['date'].dt.year
data.sample(5)
```

### 2. Extract Month

```python
data['month']=data['date'].dt.month
data.sample(5)
```

```python
data['name_of_month']=data['date'].dt.month_name()
data.sample(5)
```

### 3. Extract Date

```python
data['day']=data['date'].dt.day
data.sample(5)
```

```python
data['name_of_day']=data['date'].dt.day_name()
data.sample(5)
```

```python
data['day_of_the_week']=data['date'].dt.day_of_week+1
data.sample(5)
```

```python
data['day_of_the_year']=data['date'].dt.day_of_year
data.sample(5)
```

```python
data['is_weekend']=np.where(data['day_of_the_week'].isin([7,6]),1,0)
data.sample(5)
```

```python
data['quarter']=data['date'].dt.quarter
data['semester']=np.where(data['quarter'].isin([1,2]),1,2)

data.sample(10)
```

Extract Time elapse between dates

```python
import datetime

today=datetime.datetime.today()

today
```

```python
today-data['date'][:5]
```

Day passed

```python
(today-data['date']).dt.days
```

`Unit M is not supported. Only unambiguous timedelta values durations are supported. Allowed units are 'W', 'D', 'h', 'm', 's', 'ms', 'us', 'ns'`

```python
np.round((today-data['date'])/np.timedelta64(1,'D')/30,0)
```
