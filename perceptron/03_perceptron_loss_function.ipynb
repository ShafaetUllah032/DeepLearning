```python
# 0. import dataset form sklearn
from sklearn.datasets import make_classification

# 1. Create our features and class 
X,y=make_classification(n_classes=2,
                        n_clusters_per_class=1,
                        n_samples=100,
                        n_features=2,
                        n_informative=1,
                        n_redundant=0,
                        hypercube=False,
                        class_sep=15,
                        random_state=41)

print(f" The shape of features : {X.shape} and the shape of class : {y.shape}")
```

## 2. Data Visualization
Our data is ready , it's time to Visualize our data..

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
```

## 3. Create model with loss function

we have ploted a beautiful model, Now it's time to create our model for futher experiments

```python
def perceptron(X,y):
    w1=w2=b=1
    lr=0.1
    for j in range(100000):
        for i in range(X.shape[0]):

            # check condition
            z=w1*X[1][0] + w2*X[i][1] + b

            if z*y[i] < 0:
                w1=w1 + lr*y[i]*X[i][0]
                w2=w2 + lr*y[i]*X[i][1]
                b = b +lr*y[i]
        return w1,w2,b
```

## 4. Train Model and find weights

```python
w1,w2,b = perceptron(X,y)
```

```python
b
```

Let's find intercect and gradient 

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

## 5. Visualize the plot to see the region 

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
